```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from math import *
```

```python
# Parâmetros
c = 0.1
w = 1   # Velocidade angular em rad/s
p = 1
A =1
Cd = 0.47
m=0.5
vv = 500 # Velocidade em L/min do vento


def modelo(X,t):
    
    x = X[0] # Posição x
    y = X[1] # Posição y
    vx = X[2] # Velocidade x
    vy = X[3] # Velocidade y

    # Equações diferenciais
    dxdt = vx
    dydt = vy
    dvxdt = ( c * w ) / m
    dvydt = 0
    return [dxdt,dydt,dvxdt,dvydt]
```

```python
x_0 = 0 # posicao inicial no eixo x
y_0 = 0 # posicao inicial no eixo y
Vx_0 = 0 # velocidade inicial na direção do eixo x
Vy_0 = 0 # velocidade inicial na direção do eixo y
condicoes_iniciais = [x_0, y_0, Vx_0, Vy_0] 

tempo_inicial = 0
tempo_final = 5
dt = 1


tempo = np.arange(tempo_inicial, tempo_final, dt)

lista = odeint(modelo, condicoes_iniciais, tempo)
lista_velocidade_x = lista[:, 0]
lista_aceleracao_x = lista[:, 2]

plt.plot(tempo, lista_velocidade_x)
plt.grid()
plt.show()
```
